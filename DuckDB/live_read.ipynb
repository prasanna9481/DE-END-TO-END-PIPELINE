```python
import duckdb
import time
while True:
    try:
        # Open a read-only connection (retry on failure)
        conn = duckdb.connect('redshift_queries.duckdb', read_only=True)
        
        # Count entries
        result = conn.execute("SELECT COUNT(*) FROM live_queries;").fetchdf()
        count = result.iloc[0, 0]
        print(f"Number of entries: {count}")
        
        
    except Exception as e:
        print(f"Reader Error: {e}")
        
    finally:
        if 'conn' in locals():
            conn.close()  # Ensure connection is closed
    
    time.sleep(1)
```
